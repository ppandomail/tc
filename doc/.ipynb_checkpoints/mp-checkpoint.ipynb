# Autómatas de Células de McCulloch-Pitts
---

* Proviene del estudio de la estructura de las neuronas humanas y sus conexiones
* Constituyeron el primer paso para lo que hoy se denomina RNA
* Están constituidos por un conjunto de células interconexionadas entre sí 

## Propiedades de las células

* Pueden estar en dos estados: activo (1) o inactivo (0)
* Pueden recibir dos tipos de señales a partir de otras células o como entrada del autómata: positivas (1) o negativas(0)
* Pueden tener una o varias entradas ($e_i$) de dos tipos: exitadoras ($\rightarrow$) o inhibidoras ($\multimap$)
