# Máquinas secuenciales
---

* Representan un tipo de autómata que es capaz de, dada una palabra de entrada, generar otra palabra de salida
* Se pueden ver como un automáta qur tiene dos cintas asociadas: una por la que va leyendo las palabras de entrada, y otra de salida, en la que va generando la salida

## Máquina finita de Mealy

* Es un tipo de máquina de estados finitos
* Dada $w$ de entrada genera otra salida
* Los símbolos de salida dependen del estado y del símbolo de estrada
* Creada por George Mealy en 1955 
* Formalmente una Máquina de Mealy se define como una 5-upla: $ME = <\Sigma_E, \Sigma_S, Q, f, g>$

|            |                                                             |
| --         | --                                                          |
| $\Sigma_E$ | alfabeto de símbolos de entrada                             |
| $\Sigma_S$ | alfabeto de símbolos de salida                              |
| $Q$        | conjunto finito no vacío de estados                         |
| $f$        | función de transición, $f : Q$ x $\Sigma_E \rightarrow Q$   |
| $g$        | función de salida, $g: Q$ x $\Sigma_E \rightarrow \Sigma_S$ |

### Ejemplo: binario % 2 == 0 ? p : i

$ME_1 = <\{0, 1\}, \{p, i\}, \{q0, q1\}, f, g>$



|                 |                |
| --              | --             |
| $f(q0, 0) = q0$ | $g(q0, 0) = p$ |
| $f(q0, 1) = q1$ | $g(q0, 1) = i$ |
| $f(q1, 0) = q1$ | $g(q1, 0) = i$ |
| $f(q1, 1) = q0$ | $g(q1, 1) = p$ |

| $f$  | $0$  | $1$  |
| --   | --   | --   |
| $q0$ | $q0$ | $q1$ |
| $q1$ | $q1$ | $q0$ |

| $g$  | $0$ | $1$ |
| --   | --  | --  |
| $q0$ | $p$ | $i$ |
| $q1$ | $i$ | $p$ |

| $f/g$ | $0$    | $1$    |
| --    | --     | --     |
| $q0$  | $q0/p$ | $q1/i$ |
| $q1$  | $q1/i$ | $q0/p$ |

### Ejemplo: unario % 3

$ME_2 = <\{1\}, \{0, 1, 2\}, \{q0, q1, q2\}, f, g>$



| $f/g$ | $1$    |
| --    | --     |
| $q0$  | $q1/1$ |
| $q1$  | $q2/2$ |
| $q2$  | $q0/0$ |

### Ejemplo: $(a \mid b)^* (aa \mid bb)$

$ME_3 = <\{a, b\}, \{F, T\}, \{q0, q1, q2\}, f, g>$



| $f/g$ | $a$    | $b$    |
| --    | --     | --     |
| $q0$  | $q1/F$ | $q2/F$ |
| $q1$  | $q1/T$ | $q2/F$ |
| $q2$  | $q1/F$ | $q2/T$ |

## Máquina finita de Moore

* Es un autómata de estado finitos
* Dada $w$ de entrada genera otra salida
* La función $g$ sólo depende del estado en el que se esté
* Creado por Edward Moore en 1956
* Formalmente una Máquina de Moore se define como una 5-upla: $MO = <\Sigma_E, \Sigma_S, Q, f, g>$

|            |                                                             |
| --         | --                                                          |
| $\Sigma_E$ | alfabeto de símbolos de entrada                             |
| $\Sigma_S$ | alfabeto de símbolos de salida                              |
| $Q$        | conjunto finito no vacío de estados                         |
| $f$        | función de transición, $f : Q$ x $\Sigma_E \rightarrow Q$   |
| $g$        | función de salida, $g: Q$ \rightarrow \Sigma_S$             |

### Ejemplo: binario % 2 == 0 ? p : i

$MO_1 = <\{0, 1\}, \{p, i\}, \{q0, q1\}, f, g>$



|                 |             |
| --              | --          |
| $f(q0, 0) = q0$ | $g(q0) = p$ |
| $f(q0, 1) = q1$ |             |
| $f(q1, 0) = q1$ | $g(q1) = i$ |
| $f(q1, 1) = q0$ |             |

| $f$  | $0$  | $1$  |
| --   | --   | --   |
| $q0$ | $q0$ | $q1$ |
| $q1$ | $q1$ | $q0$ |

| $g$  | $\Sigma_S$ |
| --   | --         |
| $q0$ | $p$        |
| $q1$ | $i$        |

| $Q$    | $0$  | $1$  |
| --     | --   | --   |
| $q0/p$ | $q0$ | $q1$ |
| $q1/i$ | $q1$ | $q0$ |

### Ejemplo: binario % 3

$MO_2 = <\{0, 1\}, \{0, 1, 2\}, \{q0, q1, q2\}, f, g>$



| $Q$    | $0$  | $1$  |
| --     | --   | --   |
| $q0/0$ | $q0$ | $q1$ |
| $q1/1$ | $q2$ | $q0$ |
| $q2/2$ | $q1$ | $q2$ |

## Diferencia entre Mealy y Moore

* Las **Máquinas de Mealy** suponen velocidad de proceso **infinita**, ya que estas máquinas generan una salida inmediatamente después de recibir la entrada.
* Por el contrario, las de **Moore** suponen una velocidad **finita**, ya que la respuesta sólo depende del estado en el que se encuentre la máquina después de realizar cada transición
