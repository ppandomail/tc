# Autómatas Finitos Probabilísticos
---

* Permiten introducir la idea de que las transiciones entre estados a partir de símbolos de entrada pueden no producirse de forma segura (probabilidad igual a 1) sino que existe una determinada probabilidad asociada a que se produzca la transición
* No se puede hablar del estado en el que está el autómata en un determinado instante, sino que se habla de la probabilidad de que se encuentre en cada uno de los estados del autómata.

## Aplicaciones

* Movimientos de los robots
* Reconocimiento de voz
* Entendimiento del lenguaje natural
* Etc.

## Definición

* Formalmente un AFP se define como una 5-upla: $AFP = <\Sigma, Q, M, P(0), F>$

|          |                                                                                                 |
| --       | --                                                                                              |
| $\Sigma$ | alfabeto de símbolos de entrada                                                                 |
| $Q$      | conjunto finito no vacío de estados                                                             |
| $M$      | conjunto de matrices de probabilidad de transición entre estados, $M = \{M(a) / a \in \Sigma\}$ |
| $P(0)$   | es el vector de estado inicial                                                                  |
| $F$      | conjunto de estado finales, $F \subseteq Q$                                                     |
