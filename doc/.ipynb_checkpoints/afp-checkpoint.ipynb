# Autómatas Finitos Probabilísticos
---

* Permiten introducir la idea de que las transiciones entre estados a partir de símbolos de entrada pueden no producirse de forma segura (probabilidad igual a 1) sino que existe una determinada probabilidad asociada a que se produzca la transición
* No se puede hablar del estado en el que está el autómata en un determinado instante, sino que se habla de la probabilidad de que se encuentre en cada uno de los estados del autómata.

## Aplicaciones

* Movimientos de los robots
* Reconocimiento de voz
* Entendimiento del lenguaje natural
* Etc.

## Definición

* Formalmente un AFP se define como una 5-upla: $AFP = <\Sigma, Q, M, P(0), F>$

|          |                                                                                                 |
| --       | --                                                                                              |
| $\Sigma$ | alfabeto de símbolos de entrada                                                                 |
| $Q$      | conjunto finito no vacío de estados                                                             |
| $M$      | conjunto de matrices de probabilidad de transición entre estados, $M = \{M(a) / a \in \Sigma\}$ |
| $P(0)$   | es el vector de estado inicial                                                                  |
| $F$      | conjunto de estado finales, $F \subseteq Q$                                                     |

## Matrices de probabilidad de transición

* Reemplazan a las funciones de transición
* $\forall \alpha \in \Sigma \; \exists M(\alpha)$
* $M(\alpha)$: define la probabilidad de dado que el AFP se encuentre en un determinado estado y reciba el símbolo de entrada $\alpha$, transite a cada uno de los demás estados

$$M(\alpha) =
\begin{pmatrix}
p_{1,1} & p_{1,2} & \cdots & p_{1,n} \\
p_{2,1} & p_{2,2} & \cdots & p_{2,n} \\
\vdots & \vdots & \ddots & \vdots \\
p_{n,1} & p_{n,2} & \cdots & p_{n,n} 
\end{pmatrix}
$$

* $n$ es el número de estados, $n = |Q|$
* $p_{ij}$ es la probabilidad de que estando en el estado $i$ y recibiendo una $\alpha$ como entrada, transite al estado $j$
* $\forall p_{ij}$, se cumple $0 \leq p_{ij} \leq 1$; y
* $\forall i, \sum_{j=1}^{n} p_{ij} = 1$ 

## Ejemplo

$AFP_1 = <\{0, 1\}, \{q1, q2, q3\}, \{M(0), M(1)\}, (1/3 \; 1/3 \; 1/3), \{q3\}>$

$$M(0) =
\begin{pmatrix}
1/3 & 1/3 & 1/3 \\
1/2 & 0 & 1/2 \\
0 & 1 & 0
\end{pmatrix}
$$

$$M(1) =
\begin{pmatrix}
1/2 & 1/2 & 0 \\
1/3 & 2/3 & 0 \\
0 & 0 & 1
\end{pmatrix}
$$

* $M(0)$ y $M(1)$ definen las probabilidades de transición para los símbolos de entrada 0 y 1 respectavamente
* La prob de que, al estar en el estado q2 y recibir un 0 en la entrada, transite al estado q1 es $M_{21}(0) = 1/2$
* La prob de que, al estar en el estado q2 y recibir un 1 en la entrada, transite al estado q1 es $M_{21}(1) = 1/3$

## Vectores de estados

* El vector de estados en un instante de tiempo $t$, $P(t)$, tiene una componente por cada estado del AFP, y, el contenido de cada posición $i$ del vector, $P_i(t)$, se corresponde a la probabilidad de que en ese instante $t$ el AFP se encuentre en el estado $i$

$$P(t) = (P_1(t), P_2(t), ..., P_n(t))$$

* Se debe cumplir que $\sum_{i=1}^{n} P_i(t) = 1$ para cada instante $t$

$$P_i(t+1) = \sum_{j=1}^{n} P_j(t)M_{ji}(\alpha)$$

## Ejemplo

* Dado $AFP_1$, si el vector de estados es el inicial, $P(0) = (1/3 \; 1/3 \; 1/3)$, y se recibe un $0$, las componentes del vector de estados serán:

* $P_1(1) = \sum_{j=1}^{3} P_j(0)M_{j1}(0)$ = 1/3 x 1/3 + 1/3 x 1/2 + 1/3 x 0 = 5/18
* $P_2(1) = \sum_{j=1}^{3} P_j(0)M_{j2}(0)$ = 1/3 x 1/3 + 1/3 x 0 + 1/3 x 1 = 4/9
* $P_3(1) = \sum_{j=1}^{3} P_j(0)M_{j3}(0)$ = 1/3 x 1/3 + 1/3 x 1/2 + 1/3 x 0 = 5/18

* y $P(1)$ sería, $P(1) = (5/18 \; 4/9 \; 5/18)$ y, por tanto, sería más probable encontrarse en el estado q2.

* Si se recibiera como entrada un $1$, las componentes del vector de estados en el momento $t = 1$ serían:

* $P_1(1) = \sum_{j=1}^{3} P_j(0)M_{j1}(1)$ = 1/3 x 1/2 + 1/3 x 1/3 + 1/3 x 0 = 5/18
* $P_2(1) = \sum_{j=1}^{3} P_j(0)M_{j2}(1)$ = 1/3 x 1/2 + 1/3 x 2/3 + 1/3 x 0 = 7/18
* $P_3(1) = \sum_{j=1}^{3} P_j(0)M_{j3}(1)$ = 1/3 x 0 + 1/3 x 0 + 1/3 x 1 = 1/3

* y $P(1)$ sería, $P(1) = (5/18 \; 7/18 \; 1/3)$ 

## Lenguaje aceptado por un AFP

* Si se recibe como entrada el símbolo $a \in \Sigma$, se puede calcular el vector de estados en el instante 1, $P(1)$, a partir del vector de estados inicial $P(0)$ y de la matriz de probabilidad de transición $M(a)$ como:

$$P(1) = P(0) M(a)$$

* Si en ese instante 1 se recibe el símbolo $b \in \Sigma$, el vector de estados en el instante 2, se puede calcular análogamente con el $P(1)$ y $M(b)$

$$P(2) = P(1) M(b)$$

* Sustituyendo

$$P(2) = P(0) M(a) M(b)$$

* Generalizando

$$P(p) = P(0) M(a_1) M(a_2) ... M(a_p) = P(0) \prod_{i=1}^p M(a_i)$$ 

* En el $AFP_1$, el vector de estados (del instante 2) después de recibir la palabra de entrada w = 01

$$P(01) = P(0) M(0) M(1)$$
$$P(01) = (1/3 \; 1/3 \; 1/3) 
\begin{pmatrix}
1/3 & 1/3 & 1/3 \\
1/2 & 0 & 1/2 \\
0 & 1 & 0
\end{pmatrix} 
\begin{pmatrix}
1/2 & 1/2 & 0 \\
1/3 & 2/3 & 0 \\
0 & 0 & 1
\end{pmatrix}$$
$$P(01) = (5/18 \; 4/9 \; 5/18)
\begin{pmatrix}
1/2 & 1/2 & 0 \\
1/3 & 2/3 & 0 \\
0 & 0 & 1
\end{pmatrix}$$
$$P(01) = (31/108 \; 47/108 \; 5/18)$$

* A partir de estas definiciones, se puede calcular la probabilidad de que un determinado estado $i$ sea accesible desde q0 al recibir una determinada palabra de entrada $w \in \Sigma^*$.
* Para ello, se calcula la probabilidad de que, desde q0 se llegue al estado $i$ al realizar transiciones con $w$, por la fórmula siguiente y se seleccionar la posición $i, P_i(w)$

$$P(p) = P(0) M(a_1) M(a_2) ... M(a_p) = P(0) \prod_{i=1}^p M(a_i)$$ 

* Ahora, se puede calcular la probabilidad de que, al recibir en la entrada la palabra $w$, se llegue a un estado final, $P_F(w)$, sin más que sumar las $P_i(w)$ de los estados $i \in F$
$$P_F(w) = \sum_{i \in F} P_i(w)$$

* En el $AFP_1$, la probabilidad de que el AFP transite a un estado final desde q0 al recibir $W = 01$ será 5/18, ya que sólo hay un estado final, q3, y $P_3(01) = 5/18$

* A veces se desea saber si el AFP ha aceptado o no una determinada palabra
* Para ello, se define un umbral $\Theta$, que cumpla $0 \leq \Theta \leq 1$
* Un AFP aceptará $w$ si $P_F(w) \geq \Theta$
* Cuanto mayor sea el umbral, más restrictivo será el AFP
* Formalmente, se amplia la definición de AFP:

$$AFP = <\Sigma, Q, M, P(0), F, \Theta>$$

* El lenguaje aceptado por el AFP serán todas las palabras cuyas transiciones lleven a al gún estado final con una probabilidad mayor o igual que el umbral:

$$L_{AFP} = \{w / P_F(w) \geq \Theta, w \in \Sigma^*\}$$

* En el $AFP_1$, si se define el umbral $\Theta = 1/2$, la palabra $01$ no sería aceptadan por el AFP, ya que:

$$P_F(01) = 5/18 < 1/2$$

* Si el umbral fuera $\Theta = 1/24$, sí sería aceptada

## AF como AFP

* Los AFD y AFN son un caso particular de AFP, en el que las probabilidades con 0 ó 1
* Si se tiene un $AFD = <\Sigma, Q, q0, F, \delta>$, se puede obtener el AFP equivalente:

$$AFP = <\Sigma, Q, M, P(0), F, \Theta>$$

* donde:
    * $P(0)$ debe reflejar la idea de que sólo se está inicialmente en un estado, q0, y en ninguno más. Por tanto,
        $$
        P_i(0)=
        \begin{cases}
        1 \quad \text{si i = q0}\\ 
        0 \quad \text{en caso contrario}
        \end{cases}
        $$
    * todas las transiciones permitidas por $\delta$ tienen probabilidad igual a 1, mientras que las que no se producen, tienen probabilidad igual a 0. Por tanto, para cada símbolo $\sigma \in \Sigma$,
        $$
        M_{ij}(\sigma)=
        \begin{cases}
        1 \quad \text{si $\delta(i, \sigma) = j$}\\ 
        0 \quad \text{en caso contrario}
        \end{cases}
        $$
    * el umbral debe ser mayor que cero, siendo válido cualquier valor
